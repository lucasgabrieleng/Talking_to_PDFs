```python
# Imports
import os
import openai
import langchain
import chromadb
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_community.llms import OpenAI
from langchain.chains.question_answering import load_qa_chain
import warnings
warnings.filterwarnings('ignore')
```

## Extracting Text from PDF Files

```python
# Function to read PDF files.
def dsa_read_pdf(directory_path):
    
    # Access the folder containing the PDF file.
    file_loader = PyPDFDirectoryLoader(directory_path)
    
    # Read the document in the folder.
    documents = file_loader.load()
    
    return documents
```

```python
# Execute the function
dsa_doc = dsa_read_pdf('files/')
```

## Defining the Vector Store

https://www.trychroma.com/

```python
# Defining the index name
index_name = 'dsa-index'
```

```python
# Create the vector store.
index = Chroma.from_documents(dsa_doc, gerador_embeddings, collection_name = index_name)
```

```python
type(index)
```

```python
# Define the similarity search function
def dsa_busca_similaridade(query, k = 2):
    
    # Uses the 'similarity_search' method of the 'index' object to find the 'k' results
    # most similar to 'query'
    matching_results = index.similarity_search(query, k)
    
    # Returns the matching results of the similarity search
    return matching_results
```

## Creating the Personal Assistant App with LangChain

```python
# Create an instance of the LLM
llm_dsa = OpenAI(openai_api_key = os.environ['OPENAI_API_KEY'], temperature = 0.3)
```

https://python.langchain.com/docs/modules/chains/document/stuff

https://js.langchain.com/docs/use_cases/question_answering/

```python
# Creates a chain for questions and answers in documents.
chain = load_qa_chain(llm_dsa, chain_type = 'stuff')
```

```python
# Define the function to get the response.
def obter_resposta(query):

    # It calls the function 'dsa_busca_similaridade' with the provided query and stores the result in 'doc_search'.
    doc_search = dsa_busca_similaridade(query)
        
    # It uses the 'chain' object to execute the run function and process the 'query' and the documents found. 
    # storing the response in 'response'
    response = chain.run(input_documents = doc_search, question = query)
    
    # Returns the response obtained from the previous processing.
    return response
```

## Running the AI ​​Assistant and Conversing with PDFs

```python
# Ask the pdf file
dsa_prompt1 = "Which are the CONCERNS OVER AI IN AUTOMATION"
```

```python
# Gets the answer
answer1 = obter_resposta(dsa_prompt1)
```
